# Библиотека Pandas

Библиотека Pandas предоставляет множество операций работы с данными, обычно реализуемых электронными таблицами и базами данных.

В Pandas реализованы два основных типа данных:
    
    - Series - одномерным массив индексированных данных
    - DataFrame - многомерный массив с метками для строк и столбцов, а также, зачастую, неоднородным типом данных и пропущенными значениями
    
    

```python
import pandas as pd
```

#### Создание набора данных

```python
names = ['Петров Вася', 'Иванов Петя', 'Смирнов Ваня', 'Орлова Люба', 'Круглова Даша']
matem = [4,5,3,4,5]
inform = [5,5,4,4,5]
PupilDataSet = list(zip(names, matem, inform))
```

```python
PupilDataSet
```

Создание DataFrame на основе списка строк таблицы

```python
df = pd.DataFrame(data = PupilDataSet, columns = ['Фамилия, имя', 'математика', 'информатика'])
df
```

Создание DataFrame на основе списка словарей

```python
df1 = pd.DataFrame([{'Фамилия, имя': 'Сидоров Петя', 'математика': 5, 'информатика': 4}], index=[len(df)])
```

Добавление новых данных в таблицу

```python
df = pd.concat([df, df1])
```

```python
df
```

Генерация случайного набора данных

```python
import random
```

```python
candidates = ["Путин", 'Зюганов', 'Мишустин', 'Греф', 'Кириенко']
random.seed(500)
random_names = [candidates[random.randint(0, 4)] for i in range(10000)]
```

```python
random_names[:10]
```

```python
random_names[-5:]
```

```python
df1 = pd.DataFrame(random_names, columns = ['name'])
```

```python
df1.head(5)
```

Создание столбца с константным значением

```python
df1['Количество голосов'] = 1
```

```python
df1.tail(5)
```

#### Группировка данных

```python
name = df1.groupby('name').count()
```

```python
name
```

```python
name = df1.groupby('name').sum()
```

```python
name
```

Универсальный способ группировки с вычислением агрегатных функций

**<Имя DataFrame>.groupby(<Столбцы группировки>, as_index=False).agg({<Имя столбца>:<Имена агрегатных функций>})**

```python
df1.groupby('name', as_index=False).agg({'Количество голосов':'sum'})
```

```python
df.groupby('математика', as_index=False).agg({'Фамилия, имя':'count'})
```

```python
df.groupby(['математика', 'Сумма баллов'], as_index=False).agg({'Фамилия, имя':'count'})
```

```python
df.groupby('Сумма баллов', as_index=False).agg({'математика':['max', 'min']})
```

```python
df.groupby('Сумма баллов', as_index=False).agg({'математика':['max', 'min'], 'информатика':['max', 'min']})
```
