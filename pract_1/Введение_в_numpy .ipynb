### Одномерные и многомерные массивы

Главный объект numpy - array. Массивы схожи со списками в python, но элементы массива должны иметь одинаковый тип данных (float и int). С массивами можно проводить числовые операции с большим объемом информации в разы быстрее и, главное, намного эффективнее чем со списками.

Массивы NumPy быстрее и компактнее, чем списки Python. Массив потребляет меньше памяти и удобны в использовании. NumPy использует гораздо меньше памяти для хранения данных и предоставляет механизм указания типов данных. Это позволяет оптимизировать код еще лучше.

```python
a = np.array([1, 2, 2.7])
print(a, '|', type(a))
```

```python
help(np.array)
```

```python
a.dtype
```

Массив является центральной структурой данных библиотеки NumPy. Массив - это сетка значений, которая содержит информацию о необработанных данных, о том, как найти элемент и как интерпретировать элемент. Он имеет сетку элементов, которые могут быть проиндексированы различными способами. Все элементы имеют одинаковый тип, называемый массивом dtype. Массив может быть проиндексирован набором неотрицательных целых чисел, логическими значениями, другим массивом или целыми числами. Ранг массива - это число измерений. Форма массива представляет собой набор целых чисел, дающих размер массива по каждому измерению.

```python
a = np.array([[1, 2, 3], [4, 5, 6]], float)
print(a, '|', type(a))
```

Мы можем получить доступ к элементам в массиве, используя квадратные скобки. Когда вы обращаетесь к элементам, помните, что индексирование в NumPy начинается с 0. Это означает, что если вы хотите получить доступ к первому элементу в вашем массиве, вы получите доступ к элементу «0».

```python
a[0]
```

В двумерном массиве сначала указывем номер строки, потом номер столбца.

Нумерация строк и столбцов начинается с нуля.

Доступ к элементу во 2 строке, 3 столбце

```python
a[0][1]
```

```python
a[0,1]
```

Иногда вы можете услышать термин массив, называемый «ndarray», который является сокращением от «N-мерный массив». N-мерный массив - это просто массив с любым количеством измерений. Вы также можете сказать 1-D, или одномерный массив, 2-D, или двумерный массив, и так далее. 

Класс NumPy ndarray используется для представления как матриц, так и векторов. Вектор - это массив с одним измерением (нет разницы между векторами строки и столбца), а матрица ссылается на массив с двумя измерениями. 

Для трехмерных или более крупных размерных массивов также обычно используется термин тензор.

## Скорость операций

Рассмотрим ситуацию, в которой к каждому элементу последовательности необходимо добавить некоторую константу. Реализуем это с помощью python и numpy, сравним время вычислений.

```python
import datetime
start = datetime.datetime.now()

a = [1 for i in range(100_000_000)]
a = [elem + 10 for elem in a]

end = datetime.datetime.now()
(end-start).seconds
```

```python
import datetime
import numpy as np
start = datetime.datetime.now()

a = np.ones(100_000_000)
a = a+10

end = datetime.datetime.now()
(end-start).seconds
```

## Сравнение массивов

При сравнении массивов будут попарно сравниваться элементы, и в результате будет получен логический массив с результатами сравнения.

## Фильтрация элементов массива

Вместо индекса можно указать "маску": если элемент нужен - True, иначе - False. Будут выбраны те элементы, которым соотвествует True.

```python
a = np.array([1, 12, 33, 40, -5, -6])
maska = np.array([True, False, True, False, False, True])
a[maska]
```

Значение маски может быть получено в результате сравнения массива. Выберем все элементы, большие 5. 

```python
maska_1 = a > 5
maska_1
```

```python
a[maska_1]
```

Использовать отдельную переменную для маски необязательно.

```python
a[a > 5]
```

```python
a[a % 2 == 0]
```

Можно инвертировать маску.

```python
a[~maska_1]
```

```python
a[~(a > 5)]
```

Объединение нескольких условий логическим И.

Скобки обязательны!

```python
a[(a > 5) & (a % 2 == 0)]
```

Объединение нескольких условий логическим ИЛИ.

Скобки обязательны!

```python
a[(a > 5) | (a % 2 == 0)]
```

## Добавление и удаление в массиве

Чтобы удалить элементы из массива, просто использовать индексирование для выбора элементов, которые вы хотите сохранить.

```python
a = np.array([1, 2, 3])
# берем элементы на первом и 2 индексе
a = a[[1, 2]]
a
```

В Numpy массив можно добавлять элементы:

```python
np.append(a, np.array([4, 5, 6, 7]))
```

Массивы можно объединять

```python
a = np.array([[1, 2], [3, 4]], float)
b = np.array([[5, 6], [7,8]], float)
np.concatenate((a,b))
```

## Оси в массивах

В массивах NumPy существует понятие оси измерений. У массива столько же осей, сколько измерений. Для двумерного массива:



Если массив не одномерный, можно задать ось, по которой будет происходить соединение. По умолчанию axis=0.

```python
np.concatenate((a,b), axis=0)
```

```python
np.concatenate((a,b), axis=1)
```

```python
c = np.array([[10, 20, 30, 40], [50, 60, 70, 80]])
c
```

Считаем сумму по столбцам (двигаемся вдоль оси 0)

```python
np.sum(c, axis=0)
```

Считаем сумму по строкам (двигаемся вдоль оси 1)

```python
np.sum(c, axis=1)
```

## Работа по ссылке

Работа по ссылке как и в python:

```python
a = np.array([1, 2, 3])
b = a
c =  a.copy()
d = a[:]

# меняем 0й элемент
a[0] = 0

print(f'Измененный массив а {a}')
print(f'Измененный массив b {b}  <-- изменился по ссылке т.к. поменяли 0й элемент а')
print(f'Измененный массив c {c}  <-- создали копию в памяти')
print(f'Измененный массив d {d}  <-- изменился по ссылке т.к. поменяли 0й элемент а')
```
